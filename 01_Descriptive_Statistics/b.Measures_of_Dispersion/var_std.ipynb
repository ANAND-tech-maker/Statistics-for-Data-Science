### ðŸ“‰ Variance
#### Variance shows how far the data values are spread out from the mean.
#### It tells how much the numbers differ from the average value.
#### Formula:
#### Variance = (Sum of (x âˆ’ mean)Â²) / Number of values
#### Example:
#### Data = [2, 4, 6, 8]
#### Mean = 5
#### Variance = 5

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```
