## Shape of Distribution
The shape of a distribution describes how the data is spread.
It mainly depends on skewness, kurtosis, and symmetry.

**1. Skewness**
Skewness tells whether the data is pulled more to one side.

**Types of Skewness**
 - Positive Skew (Right Skew) – tail is longer on the right, Mean > Median
 - Negative Skew (Left Skew) – tail is longer on the left, Mean < Median
 - Zero Skew (Symmetrical) – both sides look similar, Mean ≈ Median

**Skewness (in words):**
Skewness measures how much the data is pulled toward the left or right side.


## 2. Kurtosis
Kurtosis tells how heavy the tails are and how sharp or flat the peak is.

**Types of Kurtosis**
  - Leptokurtic – heavy tails, sharp peak
  - Mesokurtic – normal distribution
  - Platykurtic – light tails, flat peak
  
**Kurtosis (in words):**
Kurtosis measures how extreme or heavy the outliers are in the data.

### Quick Interpretation Table

| Measure     | Condition      | Meaning          |
|-------------|----------------|------------------|
| Skewness    | > 0            | Right-skewed     |
| Skewness    | < 0            | Left-skewed      |
| Skewness    | = 0            | Symmetrical      |
| Kurtosis    | > 3            | Heavy tails      |
| Kurtosis    | = 3            | Normal           |
| Kurtosis    | < 3            | Light tails      |


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# ---------- SAMPLE DATA ----------
np.random.seed(42)

df = pd.DataFrame({
    "Normal_Dist": np.random.normal(loc=50, scale=10, size=500),
    "Right_Skewed": np.random.exponential(scale=20, size=500),
    "Left_Skewed": 100 - np.random.exponential(scale=20, size=500)
})

df.head()
```

```python
plt.figure(figsize=(12, 4))
sns.histplot(df["Normal_Dist"], kde=True)
plt.title("Normal Distribution")
plt.show()

plt.figure(figsize=(12, 4))
sns.histplot(df["Right_Skewed"], kde=True)
plt.title("Right-Skewed Distribution")
plt.show()

plt.figure(figsize=(12, 4))
sns.histplot(df["Left_Skewed"], kde=True)
plt.title("Left-Skewed Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df)
plt.title("Boxplots of All Distributions")
plt.show()
```

```python
print("Skewness:")
print(df.skew())
print("\nKurtosis:")
print(df.kurt())
```
