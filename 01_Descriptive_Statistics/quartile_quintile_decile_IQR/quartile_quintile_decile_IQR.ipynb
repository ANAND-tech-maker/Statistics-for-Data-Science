### Quartile Q1 (First Quartile)
#### Definition:
#### The first quartile (Q1) is the value that separates the lowest 25% of the data from the rest. It represents the 25th percentile of the dataset.
#### Formula:
$Q1 = \frac{1}{4}(n + 1)^{th} \text{ value}$


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Age': [22, 25, 30, 28, 35, 40, 27, 23, 31, 29],
    'Salary': [25000, 30000, 40000, 38000, 50000, 55000, 42000, 31000, 45000, 39000],
    'Department': ['HR', 'Finance', 'IT', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT']
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

## Q1

```python
df["Salary"].quantile(0.25) # 25th Percentile
```

### Quintile
#### Definition:
#### Quintiles divide the data into five equal parts, each containing 20% of the observations.
#### Each quintile represents a cutoff point separating 20%, 40%, 60%, 80%, and 100% of the data.
#### Formula:
$Q_k = \frac{k(n + 1)}{5} \text{th value}$
#### where
#### n = total number of observations
#### k = 1, 2, 3, 4 (for the 1st to 4th quintiles)

```python
df
```

## Q2

```python
df["Salary"].quantile(0.50)  # 50th Percentile
```

### Decile
#### Definition:
#### Deciles divide the data into ten equal parts, each representing 10% of the observations.
#### Each decile marks a cutoff point separating 10%, 20%, 30%, and so on up to 100% of the data.
#### Formula:
$D_k = \frac{k(n + 1)}{10} \text{th value}$
#### where
#### n = total number of observations
#### k = 1, 2, 3, â€¦, 9 (for the 1st to 9th decile)

```python
df
```

## Q3

```python
df["Salary"].quantile(0.75)  # 75th Percentile
```

### Interquartile Range (IQR)
#### Definition:
#### The Interquartile Range (IQR) measures the spread of the middle 50% of the data.
#### It is the difference between the third quartile (Q3) and the first quartile (Q1).
#### Formula:

$IQR = Q3 - Q1$


```python
# Q1 

df["Salary"].quantile(0.25)
```

```python
# Q3

df["Salary"].quantile(0.75)
```

```python
# IQR = Q3-Q1

df["Salary"].quantile(0.75)-df["Salary"].quantile(0.25)
```
