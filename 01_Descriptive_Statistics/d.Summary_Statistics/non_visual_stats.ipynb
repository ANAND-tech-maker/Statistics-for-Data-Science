### ðŸ“‹ Non-Visual Statistics (Numeric & Categorical)
#### Numeric Statistics
#### These are calculated for numeric columns and describe data using numbers.
#### ðŸ“˜ Common Measures:
#### Mean
#### Median
#### Variance
#### Standard Deviation
#### Minimum and Maximum

### Categorical Statistics
#### These are for text or category columns and describe frequency or uniqueness.
#### ðŸ“˜ Common Measures:
#### Count
#### unique
#### nunique
#### value_counts
#### mode

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
def num_non_visual_stats(num_df):
    '''Non-Visual Statistics for Numeric Columns'''
    for col in num_df.columns:
        print(f"{col}")
        print(num_df[col].agg(["mean", "median", "var", "std", "min", "max"]))
        print("-" * 40)
```

```python
def cat_non_visual_stats(cat_df):
    '''Non-Visual Statistics for Categorical Columns'''
    for col in cat_df.columns:
        print(f"{col}")
        print(cat_df[col].agg(["count", "nunique"]))
        print("Most Frequent Value:", cat_df[col].mode()[0])
        print("-" * 40)
```

```python
df = pd.read_excel("retail_store_sales.xlsx")
```

```python
df.head()
```

```python
numeric = df.select_dtypes(include=["int","float"])
```

```python
num_non_visual_stats(numeric)
```

```python
categorical = df.select_dtypes(include=["object"])
```

```python
cat_non_visual_stats(categorical)
```
