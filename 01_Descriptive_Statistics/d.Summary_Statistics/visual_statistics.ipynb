### ðŸ“Š Data Visualization
#### Data Visualization helps us understand data better by showing it in the form of graphs and charts.
#### It allows us to easily see patterns, trends, and relationships in the data that may not be obvious from numbers alone.
#### In this section, we will use Python libraries like:
#### Matplotlib â€“ for basic graphs (line, bar, pie, histogram)
#### Seaborn â€“ for attractive and detailed statistical visualizations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
def visual_stats(data):
    ''' Visual Analysis'''
    fig,ax = plt.subplots(2,2,figsize=(8,5),facecolor="coral",layout="constrained")
    fig.suptitle("Visual Plots")
    ax[0,0].set_title("Histogram")
    ax[0,0].set_facecolor("k")
    sns.histplot(data,ax=ax[0,0])

    ax[0,1].set_title("kde plot")
    ax[0,1].set_facecolor("yellow")
    sns.kdeplot(data,ax=ax[0,1])

    ax[1,0].set_title("ecdf plot")
    ax[1,0].set_facecolor("red")
    sns.ecdfplot(data,ax=ax[1,0])

    ax[1,1].set_title("boxpot")
    sns.boxplot(data,ax=ax[1,1])

    plt.show()

```

```python
df = pd.DataFrame({"Students":["A","B","C","D","E","F"],
             "Marks":[90,87,56,34,56,97]})
```

```python
visual_stats(df)
```
