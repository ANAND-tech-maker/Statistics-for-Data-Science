## Descriptive Statistics – Mean, Median, and Mode
#### This Jupyter Notebook focuses on the fundamental measures of central tendency — Mean, Median, and Mode.
#### These statistical metrics help summarize and understand the center or average behavior of a dataset.
#### In this notebook, we will:
#### Calculate the Mean – the average value of the data
#### Determine the Median – the middle value when data is sorted
#### Identify the Mode – the most frequently occurring value
#### These measures provide a simple yet powerful way to describe the distribution and balance of numerical data.
