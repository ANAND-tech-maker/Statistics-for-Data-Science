## ðŸ§© Types of Events in Probability  

In probability, **events** are outcomes or sets of outcomes from an experiment.  
When we consider **two or more events**, we can describe their relationship using different types of events.

---

### 1ï¸âƒ£ Mutually Exclusive Events  
- Two events are **mutually exclusive** if they **cannot occur at the same time**.  
- In other words, if one event happens, the other **cannot** happen.  

**Example:**  
When tossing a coin, getting **Head (H)** and **Tail (T)** are mutually exclusive.  
\[
P(H\T) = 0
\]

---

### 2ï¸âƒ£ Not Mutually Exclusive Events  
- Two events are **not mutually exclusive** if they **can occur together**.  

**Example:**  
When rolling a die:  
Event A = Getting an **even number**  
Event B = Getting a **number greater than 3**  
Here, both can happen together (4 and 6).  
\[
P(A \ B) not equal to 0
\]

---

### 3ï¸âƒ£ Independent Events  
- Two events are **independent** if the occurrence of one **does not affect** the other.  

**Example:**  
Tossing two different coins.  
Event A = First coin shows Head  
Event B = Second coin shows Head  
\[
P(A \ B) = P(A) * P(B)
\]

---

### 4ï¸âƒ£ Dependent Events  
- Two events are **dependent** if the outcome of one **affects** the probability of the other.  

**Example:**  
Drawing two cards **without replacement** from a deck.  
After drawing the first card, the total number of cards decreases, which changes the probability of drawing the second one.  
\[
P(A \ B) = P(A) * P(B|A)
\]


## ðŸ§® Types of Events â€” Summary Table

| Type of Event              | Description | Example | Common Outcomes | Formula |
|-----------------------------|--------------|----------|------------------|----------|
| **Mutually Exclusive** | Two events that **cannot happen together** | Rolling a die: A = odd {1,3,5}, B = even {2,4,6} | No common outcomes | \( P(A âˆ© B) = 0 \) |
| **Not Mutually Exclusive** | Two events that **can happen together** | Rolling a die: A = number > 3 {4,5,6}, B = even {2,4,6} | Common = {4,6} | \( P(A âˆ© B) not equal to 0 \) |
| **Independent Events** | One eventâ€™s outcome **does not affect** the other | Tossing two coins: A = 1st coin Head, B = 2nd coin Head | Can occur together | \( P(A âˆ© B) = P(A)* P(B) \) |
| **Dependent Events** | One eventâ€™s outcome **affects** the other | Drawing 2 cards **without replacement** | Occur together but affect each other | \( P(A âˆ© B) = P(A) * P(B\A)  |


```python
import pandas as pd

# Create a small dataset
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Fruit_1": ["Apple", "Banana", "Apple", "Mango", "Banana"],
    "Fruit_2": ["Mango", "Banana", "Apple", "Orange", "Mango"]
}

df = pd.DataFrame(data)
df
```

```python
# If Fruit_1 and Fruit_2 are same â†’ Dependent
# If different â†’ Independent
df["Event_Type"] = df.apply(lambda x: "Dependent" if x["Fruit_1"] == x["Fruit_2"] else "Independent", axis=1)
df
```

```python
# Calculate probability of each type
prob_dependent = (df["Event_Type"] == "Dependent").mean()
prob_independent = (df["Event_Type"] == "Independent").mean()

print(f"Probability of Dependent Event: {prob_dependent*100:.0f}%")
print(f"Probability of Independent Event: {prob_independent*100:.0f}%")
```

```python
# Create a dataset
data = {
    "Weather": ["Sunny", "Rainy", "Sunny", "Rainy", "Sunny", "Cloudy", "Rainy", "Cloudy", "Sunny", "Rainy"],
    "Play_Cricket": ["Yes", "No", "Yes", "No", "Yes", "Yes", "No", "Yes", "Yes", "No"],
    "Umbrella": ["No", "Yes", "No", "Yes", "No", "No", "Yes", "No", "No", "Yes"],
    "Windy": ["No", "Yes", "No", "No", "No", "Yes", "Yes", "No", "No", "Yes"]
}

df = pd.DataFrame(data)

df
```

```python
p_windy = (df["Windy"] == "Yes").mean()
p_play = (df["Play_Cricket"] == "Yes").mean()
p_both = ((df["Windy"] == "Yes") & (df["Play_Cricket"] == "Yes")).mean()

print(f"P(Windy): {p_windy*100:.0f}%")
print(f"P(Play_Cricket): {p_play*100:.0f}%")
print(f"P(Windy âˆ© Play_Cricket): {p_both*100:.0f}%")
print(f"P(Windy) and P(Play_Cricket): {(p_windy * p_play)*100:.0f}%")
```

```python
p_umbrella = (df["Umbrella"] == "Yes").mean()
p_umbrella_given_rainy = (df[df["Weather"] == "Rainy"]["Umbrella"] == "Yes").mean()

print(f"P(Umbrella): {p_umbrella*100:.0f}%")
print(f"P(Umbrella and Rainy):{p_umbrella_given_rainy*100:.0f}%")
```

```python
# Sample space for one die
sample_space = [1, 2, 3, 4, 5, 6]
n = len(sample_space)

print("Sample Space:", sample_space)
print("Total Outcomes:", n)
```

```python
# Simple Event â†’ Getting 3
A = [3]
p_A = len(A) / n
print(f"\nP(A) - Getting 3: {p_A*100:.0f}%")
```

```python
# Compound Event â†’ Getting an even number
B = [2, 4, 6]
p_B = len(B) / n
print(f"P(B) - Even number: {p_B*100:.0f}%")
```

```python
# Mutually Exclusive Events â†’ Getting 2 or 5
C = [2]
D = [5]
p_C_and_D = len(set(C).intersection(D)) / n
print(f"P(C âˆ© D) - Mutually Exclusive:  {p_C_and_D*100:.0f}%")
```

```python
# Not Mutually Exclusive â†’ Even number and number >3
# sample space = [1,2,3,4,5,6]
E = [2, 4, 6]
F = [4, 5, 6]
p_E_and_F = len(set(E).intersection(F)) / n
print(f"P(E âˆ© F) - Not Mutually Exclusive: {p_E_and_F*100:.0f}%")
```

```python
# Independent Events â†’ Two dice rolled
# Create DataFrame manually
die1 = [1, 2, 3, 4, 5, 6]
die2 = [1, 2, 3, 4, 5, 6]
df = pd.DataFrame([(x, y) for x in die1 for y in die2], columns=["Die1", "Die2"])
df.head(10)
```

```python
# Event A = 3 on first die
# Event B = 6 on second die
df["A"] = df["Die1"] == 3
df["B"] = df["Die2"] == 6
```

```python
df.head(10)
```

```python
p_A1 = df["A"].mean()
p_B1 = df["B"].mean()
p_A1_and_B1 = (df["A"] & df["B"]).mean()

df["result"] = df["A"] & df["B"]

df.tail()
```

```python
print("\nIndependent Events (Two Dice):")
print(f"P(A) = {p_A1*100:.0f}%")
print(f"P(B) = {p_B1*100:.0f}%")
print(f"P(A âˆ© B) = {p_A1_and_B1*100:.0f}%")
print(f"P(A)*P(B) = {(p_A1 * p_B1)*100:.0f}%")
```

```python
# Bag with balls
balls = ["Red", "Red", "Blue", "Blue"]

# Create all possible 2-ball draws without replacement
draws = []
for i in range(len(balls)):
    for j in range(len(balls)):
        if i != j:  # without replacement
            draws.append((balls[i], balls[j]))

df = pd.DataFrame(draws, columns=["First_Draw", "Second_Draw"])
print("All possible draws (without replacement):")
df
```

```python
# Event A: First ball is Red
df["A"] = df["First_Draw"] == "Red"

# Event B: Second ball is Blue
df["B"] = df["Second_Draw"] == "Blue"

df
```

```python
# Probability of A
p_A = df["A"].mean()

# Probability of B given A (dependent)
p_B_given_A = df[df["A"]]["B"].mean()

# Probability of A and B
p_A_and_B = p_A * p_B_given_A

print("\nDependent Events Probabilities:")
print("P(A) - First Red =", p_A)
print("P(B|A) - Second Blue given first Red =", p_B_given_A)
print("P(A âˆ© B) =", p_A_and_B)
```
