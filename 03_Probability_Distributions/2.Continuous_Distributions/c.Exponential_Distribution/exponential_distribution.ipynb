## ðŸ“Œ Exponential Distribution â€” Definition

The **Exponential Distribution** is a continuous probability distribution used to  
model the **time between independent events** that happen at a **constant average rate**.

It is commonly used for:
- Time until a machine fails  
- Time between customer arrivals  
- Time between earthquakes  
- Time required to complete a task  

### Key Characteristics
- Always **positive** (X > 0)
- **Right-skewed**
- Memoryless property:  
  P(X > s + t | X > s) = P(X > t)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_rate = 0.5
scale = 1 / lambda_rate


```

```python
# Generate sample data
data = np.random.exponential(scale, 1000)

# Plot histogram
plt.hist(data, bins=30, density=True)
```

```python

# PDF curve
x = np.linspace(0, max(data), 200)
pdf = lambda_rate * np.exp(-lambda_rate * x)
plt.plot(x, pdf)

plt.title("Exponential Distribution Visualization")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()
```
