### Pareto Distribution â€“ Definition

The **Pareto distribution** is a continuous probability distribution that describes situations where a small number of events contribute to most of the effect. It is often called the **80/20 rule distribution**, meaning **20% of the population holds 80% of the outcome**.  
A random variable \( X \) follows a Pareto distribution if it takes values \( x \ge x_m \), where \( x_m \) is the minimum value.

---

### Applications of Pareto Distribution

- **Income and wealth distribution** (a small group holds most of the wealth)
- **City population sizes**
- **File size distribution in internet traffic**
- **Business revenue distribution** (few products generate most sales)
- **Insurance and risk modeling**
- **Modeling failure rates for machines or components**


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 3      # shape parameter
xm = 1         # minimum value

# Generate Pareto data
data = (np.random.pareto(alpha, 2000) + 1) * xm

# Plot histogram
plt.hist(data, bins=30, density=True)

# Theoretical PDF
x = np.linspace(xm, max(data), 200)
pdf = (alpha * xm**alpha) / (x**(alpha + 1))
plt.plot(x, pdf)

plt.title("Pareto Distribution Visualization")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
