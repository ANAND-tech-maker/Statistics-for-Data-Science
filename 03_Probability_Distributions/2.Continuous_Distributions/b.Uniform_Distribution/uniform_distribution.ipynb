# ðŸ“˜ Uniform Distribution

The **Uniform Distribution** is a continuous probability distribution where **all values in an interval [a, b] are equally likely** to occur.

---

## âœ… Key Points

- Every value between **a and b** has the **same probability density**.  
- No value is more likely than another.  
- It is used when the outcome is completely random within a range.

---

## ðŸ“Œ Probability Density Function (in words)

The probability density at any value **x** is:

> f(x) = 1 / (b âˆ’ a), for a â‰¤ x â‰¤ b  
> f(x) = 0, otherwise

---

## ðŸŸ¢ Real-life Examples
- Random time a bus arrives between 9:00â€“9:30  
- Random number generator between 1 and 100  
- Choosing a random point on a line segment  

---

## ðŸŽ¯ When to Use
Use the Uniform Distribution when **every value in an interval is equally likely**.


```python
# ðŸ“Œ Uniform Distribution Example

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters
a = 2   # minimum value
b = 10  # maximum value

# Create x values
x = np.linspace(a - 2, b + 2, 500)

```

```python
# PDF (Probability Density Function)
pdf = uniform.pdf(x, loc=a, scale=b - a)

plt.plot(x, pdf)
plt.title("Uniform Distribution PDF")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()

# Simulating uniform random numbers
samples = np.random.uniform(a, b, 1000)

plt.hist(samples, bins=20, edgecolor='black')
plt.title("Uniform Distribution (Simulated Data)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

print("Random Uniform Samples (first 10):", samples[:10])
```
