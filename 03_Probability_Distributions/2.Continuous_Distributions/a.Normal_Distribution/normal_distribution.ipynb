# ðŸ“˜ Normal Distribution (Gaussian Distribution)

The **Normal Distribution** is a continuous probability distribution that forms a **bell-shaped curve**.  
Most values cluster around the **mean**, and the probability decreases symmetrically as you move away.

- Symmetric around the **mean (Î¼)**
- Spread is controlled by **standard deviation (Ïƒ)**

## ðŸŽ¯ When to Use
Use the normal distribution when your data forms a **bell curve**, is **continuous**, and naturally clusters around the mean.


```python
#Normal Distribution Example

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 50   # mean
sigma = 10  # standard deviation

# x values for plotting
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
```

```python
# PDF (Probability Density Function)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf)
plt.title(f"Normal Distribution (Î¼={mu}, Ïƒ={sigma})")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid()
plt.show()

```

```python
# Generate random normal values
samples = np.random.normal(mu, sigma, 1000)

plt.hist(samples, bins=30, density=True, edgecolor="black")
plt.title("Histogram of Normal Distribution Samples")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

print("First 10 samples:", samples[:10])
```
