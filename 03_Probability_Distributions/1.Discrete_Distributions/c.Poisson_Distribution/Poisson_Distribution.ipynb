# ðŸ“Š Poisson Distribution

The **Poisson Distribution** is a discrete probability distribution that models the number of events occurring in a **fixed interval of time or space**, given that these events occur with a **known constant mean rate** and **independently** of the time since the last event.

---

## âœ… Key Points

- **Random Variable (X):** Number of events in the interval  
- **Parameter (Î»):** Average number of events in the interval  
- **Discrete Distribution:** X can take values 0, 1, 2, â€¦  

---

## ðŸ“Œ Probability Mass Function (PMF) â€“ in words

The probability of observing **k events** in a fixed interval is:

> P(X = k) = (Î»^k Ã— e^(-Î»)) / k!

Where:  
- Î» = expected number of events  
- k! = factorial of k  
- e â‰ˆ 2.71828  

---

## ðŸŸ¢ Real-world Examples

1. Number of phone calls received by a call center in an hour  
2. Number of decay events per second from a radioactive source  
3. Number of accidents at a traffic intersection per month  

---

## ðŸŽ¯ When to Use

Use the **Poisson Distribution** when:  
- Events occur randomly and independently  
- The average rate (Î») is constant  
- The number of events can be 0, 1, 2, â€¦ with no upper limit


```python
# Poisson Distribution Example

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters

lam = 3  # average number of events per interval (Î»)


# Possible number of events (k)

k_values = np.arange(0, 11)  # 0 to 10 events
pmf_values = poisson.pmf(k_values, lam)

print("Number of Events (k):", k_values)
print("PMF Values:", pmf_values)
```

```python

# Visualization

plt.bar(k_values, pmf_values, color="skyblue", edgecolor="black")
plt.title(f"Poisson Distribution (Î»={lam})")
plt.xlabel("Number of Events (k)")
plt.ylabel("Probability P(X=k)")
plt.xticks(k_values)
plt.show()

```

```python

# Simulating Poisson Trials

simulated = np.random.poisson(lam, 1000)

plt.hist(simulated, bins=np.arange(0, max(simulated)+1)-0.5, rwidth=0.8, color="orange", edgecolor="black")
plt.title(f"Simulated Poisson Trials (Î»={lam}, 1000 trials)")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()

```

```python

# Example: Probability of exactly 4 events

k = 4
p_k = poisson.pmf(k, lam)
print(f"\nProbability of exactly {k} events: P(X={k}) =", p_k)
```
