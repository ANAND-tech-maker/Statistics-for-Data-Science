# üìò Binomial Distribution

The **Binomial Distribution** is a discrete probability distribution that models the number of **successes in a fixed number of independent Bernoulli trials**.

---

## ‚úÖ Key Points

- **Trials (n):** The number of repeated experiments.  
- **Success (p):** Probability of success in a single trial.  
- **Failures (q):** Probability of failure in a single trial (q = 1 - p).  
- **Random Variable (X):** Number of successes in n trials.

---

## üìå Probability Mass Function (PMF) ‚Äì in words

The probability of getting exactly **k successes** in **n trials** is:

> P(X = k) = Number of ways to choose k successes √ó (Probability of success)^k √ó (Probability of failure)^(n-k)

Where:  
- ‚ÄúNumber of ways to choose k successes‚Äù is calculated using combinations.  

---

## üü¢ Real-world Examples

1. Tossing a coin 5 times and counting heads.  
2. Number of students passing an exam out of 10.  
3. Number of defective items in a batch of 20 manufactured products.  

---

## üéØ When to Use

Use the **Binomial Distribution** when:
- There is a fixed number of independent trials (n)  
- Each trial has only two outcomes: success or failure  
- Probability of success (p) is constant across trials


```python
#  Binomial Distribution Example

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 5       # number of trials
p = 0.6     # probability of success

# Example: Probability of k successes

k_values = np.arange(0, n+1)  # possible number of successes
pmf_values = binom.pmf(k_values, n, p)

print("Number of Successes (k):", k_values)
print("PMF Values:", pmf_values)
```

```python

# Visualization

plt.bar(k_values, pmf_values, color="skyblue", edgecolor="black")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.xlabel("Number of Successes (k)")
plt.ylabel("Probability P(X=k)")
plt.xticks(k_values)
plt.show()
```

```python

# Simulating Binomial Trials

simulated = np.random.binomial(n, p, 1000)

plt.hist(simulated, bins=np.arange(-0.5, n+1.5, 1), rwidth=0.8, color="orange", edgecolor="black")
plt.title(f"Simulated Binomial Trials (n={n}, p={p}, 1000 trials)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.xticks(k_values)
plt.show()
```

```python

# Example: Probability of exactly 3 successes

k = 3
p_k = binom.pmf(k, n, p)
print(f"\nProbability of exactly {k} successes: P(X={k}) =", p_k)
```
